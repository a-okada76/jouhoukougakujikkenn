#16. ファイルをN分割する
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．


```python
import sys
from google.colab import files

N = input()
with open("popular-names.txt", 'r') as f:
    l = [] 

    for line in f:
        l.append(line)
count = 0       
for i in range(100):
    print(l[i], end='') 
    count += 1
    if count == int(N): #コマンドライン引数で受け取る場合は if count == int(sys.argv[1]): とコードを書けば良い
        print()
        count = 0
print('\n\n')

!split -l 15 popular-names.txt s_popular | cat s_popular* | head -n 100
```

# 17. １列目の文字列の異なり
1列目の文字列の種類（異なる文字列の集合）を求めよ．確認にはcut, sort, uniqコマンドを用いよ．


```python
def is_onajimoji(string_set, data):
    for i in range(len(string_set)):
        if data == string_set[i]:
            return True
    return False    

data = []
for i in range(len(l)):
    row = l[i].split()
    data.append(row)
string_set = []
for i in range(len(data)):
    if not is_onajimoji(string_set, data[i][0]):
        string_set.append(data[i][0])

for i in range(len(string_set)):
    print(string_set[i])

print(len(string_set))  
print('\n\n') 

!cut -f 1 --delim="	" popular-names.txt | sort | uniq
```

# 18. 各行を3コラム目の数値の降順にソート

```python
sorted_data = sorted(data, key=lambda x:int(x[2]), reverse=True)

for i in range(100):
    print(sorted_data[i])
print('\n\n')

!sort -k 3nr -t "	" popular-names.txt | head -n 100  
```
